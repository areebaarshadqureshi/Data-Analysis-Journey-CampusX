```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### Colored Scatterplots

```python
iris = pd.read_csv('iris.csv')
iris.sample(5)
```

```python
iris['Species'] = iris['Species'].replace({'Iris-setosa':0,'Iris-versicolor':1,'Iris-virginica':2})
iris.sample(5)
```

```python
plt.scatter(iris['SepalLengthCm'],iris['PetalLengthCm'],c=iris['Species'])
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
```

```python
plt.scatter(iris['SepalLengthCm'],iris['PetalLengthCm'],c=iris['Species'],cmap="jet")
plt.xlabel('Sepal Length')##cmap=color map
plt.ylabel('Petal Length')
```

```python
plt.scatter(iris['SepalLengthCm'],iris['PetalLengthCm'],c=iris['Species'],cmap="jet",alpha=0.7)
plt.xlabel('Sepal Length')##cmap=color map
plt.ylabel('Petal Length')##alpha;color reduce
plt.colorbar()
```

# plot size

```python
plt.figure(figsize=(15,7))

plt.scatter(iris['SepalLengthCm'],iris['PetalLengthCm'],c=iris['Species'],cmap='jet',alpha=0.7)
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.colorbar()
```

### Annotations

```python
batters = pd.read_csv('batter.csv')
```

```python
sample_df = batters.head(100).sample(25,random_state=5)
sample_df.head()
```

```python
plt.scatter(sample_df['avg'],sample_df['strike_rate'])
plt.xlabel('Weight')
plt.ylabel('Height')
```

```python
#annotation
x = [1,2,3,4]
y = [5,6,7,8]

plt.scatter(x,y)
plt.text(1,5,'Point 1')
plt.text(2,6,'Point 2')
plt.text(3,7,'Point 3')
plt.text(4,8,'Point 4',fontdict={'size':12,'color':'brown'})
```

```python
plt.figure(figsize=(18,10))
plt.scatter(sample_df['avg'],sample_df['strike_rate'],s=sample_df["runs"])

for i in range(sample_df.shape[0]):
  plt.text(sample_df['avg'].values[i],sample_df['strike_rate'].values[i],sample_df['batter'].values[i])
```

### Horizontal and Vertical lines


```python
##predicting on batsmans which has strike rate >130
plt.figure(figsize=(18,10))
plt.scatter(sample_df['avg'],sample_df['strike_rate'],s=sample_df["runs"])
plt.axhline(130,color='r',linestyle='--')
for i in range(sample_df.shape[0]):
  plt.text(sample_df['avg'].values[i],sample_df['strike_rate'].values[i],sample_df['batter'].values[i])
```

```python
##predicting on batsmans which has strike rate >130 and avg is greater than 30
plt.figure(figsize=(18,10))
plt.scatter(sample_df['avg'],sample_df['strike_rate'],s=sample_df["runs"])
plt.axhline(130,color='r',linestyle='--')
plt.axhline(140,color='g',linestyle='--')
plt.axvline(30,color='r',linestyle='--')
for i in range(sample_df.shape[0]):
  plt.text(sample_df['avg'].values[i],sample_df['strike_rate'].values[i],sample_df['batter'].values[i])
```

### Subplots

```python
# A diff way to plot graphs
batters.head()
```

```python
#first solution
plt.figure(figsize=(15,6))
plt.scatter(batters['avg'],batters['strike_rate'])
plt.title('Something')
plt.xlabel('Avg')
plt.ylabel('Strike Rate')

plt.show()
```

```python
##object and axes
fig,ax=plt.subplots(figsize=(15,6))

ax.scatter(batters['avg'],batters['strike_rate'])
ax.set_title('Something')
ax.set_xlabel('Avg')
ax.set_ylabel('Strike Rate')

fig.show()
```

```python
##two graphs avg vs runs and avg vs strike rate
fig,ax=plt.subplots(nrows=2,ncols=1,sharex=True,figsize=(15,6))
ax[0].scatter(batters['avg'],batters['runs'],color='red')
ax[1].scatter(batters['avg'],batters['strike_rate'])

ax[0].set_title('Avg Vs Strike Rate')
ax[0].set_ylabel('Strike Rate')


ax[1].set_title('Avg Vs Runs')
ax[1].set_ylabel('Runs')
ax[1].set_xlabel('Avg')
```

```python
fig,ax=plt.subplots(nrows=2,ncols=2,figsize=(15,6))
ax[0,0].scatter(batters['avg'],batters['runs'],color='red')
ax[0,1].scatter(batters['avg'],batters['strike_rate'])
ax[1,0].hist(batters['avg'])
ax[1,1].hist(batters['runs'])

```

```python
##3rd solution
fig = plt.figure()

ax1 = fig.add_subplot(2,2,1) ##(row,col,first one)
ax1.scatter(batters['avg'],batters['strike_rate'],color='red')

ax2 = fig.add_subplot(2,2,2) ##(row,col,2nd one)
ax2.hist(batters['runs'])

ax3 = fig.add_subplot(2,2,3)
ax3.hist(batters['avg'])
```

### 3D Surface Plots

```python
x = np.linspace(-10,10,100)
y = np.linspace(-10,10,100)
```

```python
#making two grids
xx,yy=np.meshgrid(x,y)
```

```python
z=xx**2+yy**2
```

```python
fig=plt.figure()
ax=fig.add_subplot(projection='3d')
p=ax.plot_surface(xx,yy,z,cmap='jet')
fig.colorbar(p)
```

```python
z = np.sin(xx) + np.cos(yy)

fig = plt.figure(figsize=(12,8))

ax = plt.subplot(projection='3d')

p = ax.plot_surface(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```

```python
z = np.sin(xx) + np.log(yy)

fig = plt.figure(figsize=(12,8))

ax = plt.subplot(projection='3d')

p = ax.plot_surface(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```

### Contour Plots
- present 3D graphs into 2D

```python
fig = plt.figure(figsize=(12,8))

ax = plt.subplot()

p = ax.contour(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```

```python
##filled contoure graph
fig = plt.figure(figsize=(12,8))

ax = plt.subplot()

p = ax.contourf(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```
